```python
params = ['run', 'test', 'msg_len', 'workers', 'pps', 'duration',]
```

```python
cpu = pd.read_csv(f"{resdir}/cpu.csv")
cpu = cpu[cpu.core == -1].reset_index(drop=True)
cpu['used'] = 100 - cpu['idle']
cpu = cpu.groupby(params).mean().reset_index()
```

```python
names = ['run', 'test', 'msg_len', 'workers', 'pps', 'duration', 'total', 'throughput_pps', 'throughput_mbps']
throughput = pd.read_csv(f"{resdir}/results.csv", sep=';', names=names, skiprows=1)
throughput['throughput_mbps'] *= 8
```

```python
throughput['expected_mbps'] = (throughput['pps'] * throughput['msg_len'] * 8)/1e6
```

```python
# throughput = throughput.set_index(params)
```

```python
df = throughput.reset_index()
sns.relplot(df, kind='line', x='msg_len', y='throughput_mbps', hue='test', hue_order=['encryption', 'decryption'], col='pps', row='workers')
```

```python
df = cpu.reset_index()
sns.relplot(df, kind='line', x='msg_len', y='used', hue='test', hue_order=['encryption', 'decryption'], col='pps', row='workers')
```

```python
diff = 'msg_len'
base = 1500
othr = 50
metric = 'used'
df = cpu
new_params = [i for i in params if i != diff]

df_base = df[df[diff] == base].set_index(new_params)
df_othr = df[df[diff] == othr].set_index(new_params)

df_diff = df_base[metric] - df_othr[metric]
df_diff = df_diff.reset_index()
df_diff_relative = (df_othr[metric] - df_base[metric]) * 100 / df_base[metric]
df_diff_relative = df_diff_relative.reset_index()


# sns.relplot(df, kind='line', x='msg_len', y=metric, col='pps', row='workers')
```
